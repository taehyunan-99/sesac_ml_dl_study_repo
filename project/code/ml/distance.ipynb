# 다양한 거리 계산

```python
import pandas as pd
from plt_rcs import *
import numpy as np
```

```python
df_dist = pd.DataFrame(
    data=[[1, 1], [4, 5], [3, 2]],
    index=['A', 'B', 'C'],
    columns=['X', 'Y']
)
```

```python
df_dist
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(
    data=df_dist, x='X', y='Y',
    s=100, fc='0.8', ec='0'
)
plt.xlim(0, 6)
plt.ylim(0, 6)
for i, row in df_dist.iterrows():
    plt.text(x=row['X'] + 0.2, y=row['Y'], s=i, 
            ha='left', va='center', fontsize=10, color='blue', fontweight='bold')
plt.grid(linestyle='--', linewidth=0.5)
plt.show()
```

## 거리 계산 관련 함수

```python
from sklearn.metrics import DistanceMetric
```

```python
DistanceMetric.get_metric(metric='manhattan').pairwise(X=df_dist)
```

```python
DistanceMetric.get_metric(metric='euclidean').pairwise(X=df_dist)
```

```python
DistanceMetric.get_metric(metric='chebyshev').pairwise(X=df_dist)
```

```python
DistanceMetric.get_metric(metric='minkowski', p=np.inf).pairwise(X=df_dist)
```

```python
cust = pd.DataFrame(data={
    '성별': [0, 0, 1],
    '학력': [1, 2, 3],
    '나이': [35, 39, 42],
    '키': [183, 179, 165],
    '몸무게': [78, 82, 50],
    '연봉': [2900, 8500, 9800]
}, index=['A', 'B', 'C'])
```

```python
cust
```

```python
DistanceMetric.get_metric(metric='euclidean').pairwise(X=cust)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(X=cust)
```

```python
scaled = pd.DataFrame(data=scaled, columns=cust.columns, index=cust.index)
```

```python
scaled.round(3)
```

```python
DistanceMetric.get_metric(metric='euclidean').pairwise(X=scaled)
```
