## OOB 결정계수 시각화

```python
def oob_score(ntree):
    model.set_params(n_estimators=ntree)
    model.fit(X_train, y_train)
    return model.oob_score_
```

```python
ntrees = range(1, 101)
oob_rsq = [oob_score(ntree) for ntree in ntrees]
```

```python
sns.lineplot(x=ntrees, y=oob_rsq, color='red', linewidth=1)
```
