```python
param_dist = {
    'max_depth': range(1, 11, 1), 
    'min_child_weight': range(1, 8, 1), 
    'learning_rate': np.round(np.arange(0.01, 0.21, 0.01), 2), 
    'colsample_bytree': np.round(np.arange(0.5, 1.01, 0.1), 1)
}
```

```python
from optuna.distributions import IntDistribution, FloatDistribution
```

```python
param_dist = {
    'max_depth': IntDistribution(low = 1, high = 10), 
    'min_child_weight': IntDistribution(low = 1, high = 7), 
    'learning_rate': FloatDistribution(low = 0.01, high = 0.20), 
    'colsample_bytree': FloatDistribution(low = 0.5, high = 1.0)
}
```

```python
from optuna.integration import OptunaSearchCV
```

```python
bo_cv = OptunaSearchCV(estimator = model, 
                       param_distributions = param_dist, 
                       n_trials = 100, 
                       timeout = 600, 
                       random_state = 0)
```

```python
bo_cv.fit(X = X_tr, y = y_tr, eval_set = [(X_vl, y_vl)], verbose = False)
```

```python
bo_cv.best_params_
```

```python
bo_cv.best_estimator_.score(X = X_tr, y = y_tr)
```

```python
bo_cv.best_estimator_.score(X = X_vl, y = y_vl)
```

```python
y_pred_grid = gs_cv.best_estimator_.predict(X = X_vl)
y_pred_rand = rs_cv.best_estimator_.predict(X = X_vl)
y_pred_optuna = bo_cv.best_estimator_.predict(X = X_vl)
```
