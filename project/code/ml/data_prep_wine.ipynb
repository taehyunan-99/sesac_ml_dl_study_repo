# 와인 분석

```python
import os
import numpy as np
import pandas as pd
from plt_rcs import *
import hds
```

```python
df = pd.read_csv('https://bit.ly/WhiteWine_Data', sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe().round(3)
```

```python
df['quality'].value_counts().sort_index()
# quality
# 3      20
# 4     163
# 5    1457
# 6    2198
# 7     880
# 8     175
# 9       5
# Name: count, dtype: int64
```

```python
df.columns
```

```python
# 서열형 -> 명목형
df['grade'] = np.where(df['quality'].ge(7), 1, 0)
```

```python
# 불필요한 컬럼 제거
df = df.drop(columns=['free sulfur dioxide', 'quality'])
```

```python
df.head()
```

## 목표변수 도수 확인

```python
plt.rc(group='figure', figsize=(4, 4))
```

```python
hds.plot.bar_freq(data=df, x='grade')
```

## 연속형 변수 간 상관관계 확인

```python
hds.plot.corr_heatmap(data=df, fontsize=7)
```

## 입력 변수와 관계 확인

```python
for var_name in df.columns.to_list()[-1]:
    hds.plot.box_group(data=df, x='grade', y=var_name, palette=['skyblue', 'orange'])
    plt.show()
```

## 특성 행렬과 타겟 벡터로 분리

```python
yvar = 'grade'
X = df.drop(columns=yvar)
y = df[yvar].copy()
display(X)
display(y)
```

## 실습 데이터셋 분할

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
# (3428, 10)
X_valid.shape
# (1470, 10)
```

```python
# 타겟 데이터의 상대도수 확인
y_train.value_counts(normalize=True).sort_index()
# grade
# 0    0.782089
# 1    0.217911
# Name: proportion, dtype: float64
y_valid.value_counts(normalize=True).sort_index()
# grade
# 0    0.787075
# 1    0.212925
# Name: proportion, dtype: float64
```
