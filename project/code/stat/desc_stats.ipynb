```python
import os
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels import robust
```

```python
df = pd.read_csv('https://bit.ly/UsedCarsPrice')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# 컬럼 타입 변경
df.columns
cols = ['MetColor', 'Automatic']
df[cols] = df[cols].astype(str)
df.dtypes
```

## 평균

```python
df['Price'].mean()
# np.float64(9690.232941176471)
```

```python
nums = pd.Series([1, np.nan, 2])
nums.mean()
# np.float64(1.5)
```

## 절사평균

```python
stats.trim_mean(df['Price'], proportiontocut=0.1)
# np.float64(9584.380019588638)
```

## 중위수

```python
df['Price'].median()
# np.float64(9450.0)
```

## 최빈값

```python
df['FuelType'].mode()
# 0    Petrol
# Name: FuelType, dtype: object
```

```python
df['FuelType'].value_counts()
# FuelType
# Petrol    1129
# Diesel     129
# CNG         17
# Name: count, dtype: int64
```

```python
# 상대도수 확인
df['FuelType'].value_counts(normalize=True)
# FuelType
# Petrol    0.885490
# Diesel    0.101176
# CNG       0.013333
# Name: proportion, dtype: float64
```

## 분위수

```python
# 4분위수
df['Price'].quantile(np.linspace(0, 1, 5))
# 0.00     4350.0
# 0.25     8250.0
# 0.50     9450.0
# 0.75    10950.0
# 1.00    15950.0
# Name: Price, dtype: float64
```

```python
# 10분위수
df['Price'].quantile(np.linspace(0, 1, 11))
# 0.0     4350.0
# 0.1     7250.0
# 0.2     7950.0
# 0.3     8500.0
# 0.4     8950.0
# 0.5     9450.0
# 0.6     9950.0
# 0.7    10500.0
# 0.8    11456.0
# 0.9    12500.0
# 1.0    15950.0
# Name: Price, dtype: float64
```

## 최솟값과 최댓값

```python
df['Price'].min()
# np.int64(4350)

df['Price'].max()
# np.int64(15950)
```

```python
df['Price'].quantile([0, 1])
# 0.0     4350.0
# 1.0    15950.0
# Name: Price, dtype: float64
```

```python
df['Price'].agg(func=['min', 'max'])
# min     4350
# max    15950
# Name: Price, dtype: int64
```

## 범위와 사분범위

```python
df['Price'].max() - df['Price'].min()
# np.int64(11600)
```

```python
df['Price'].quantile([0, 1]).diff().iloc[-1]
# np.float64(11600.0)
```

```python
# 사분위수 범위
df['Price'].quantile([0.25, 0.75]).diff().iloc[-1]
# np.float64(2700.0)
```

## 분산

- ddof는 degree of freedom이며, 1을 지정하면 편차 제곱합을 n-1로 나눔

```python
df['Price'].var()
# np.float64(4120265.326386555)
df['Price'].var(ddof=0)
# np.float64(4117033.7457384085)
```

```python
# 넘파이는 기본값 = 0
np.var(df['Price'])
# np.float64(4117033.7457384085)
```
