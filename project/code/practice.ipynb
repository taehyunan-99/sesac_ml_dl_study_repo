## 색상 통일

```python
df['색상'].unique()
```

```python
color_g = ['은색', '은색투톤', '쥐색', '명은색', '은회색', '연금색', '회색']
color_w = ['흰색', '흰색투톤']
color_b = ['검정투톤', '검정색', '은하색']
color_c = ['담녹색', '진주색', '노란색', '청색', '빨간색', '진주투톤', '하늘색', '금색', '분홍색', '주황색', '갈대색', '청옥색', '갈색', '자주색', '연두색', '보라색', '녹색', '갈색투톤', '기타']
```

```python
df.loc[df['색상'].isin(color_g), '색상'] = '은색'
df.loc[df['색상'].isin(color_w), '색상'] = '흰색'
df.loc[df['색상'].isin(color_b), '색상'] = '검정'
df.loc[df['색상'].isin(color_c), '색상'] = '기타'
```

## 사고여부 확인

```python
cols = ['가격', '제조사', '모델', '주행거리', '변속기', '연료', '튜닝여부',
        '사고여부', '내차피해', '타차가해', '연식개월', '차번변경', '차주변경', '침수여부']
```

```python
df[cols].head()
```

```python
# 사고기록('사고여부', '내차피해', '타차가해')을 하나의 '사고횟수' 컬럼으로 통합
df['사고횟수'] = df['내차피해'] + df['타차가해']
df.head()
```

```python
# 기존 사고기록('사고여부', '내차피해', '타차가해') 컬럼 제거
df = df.drop(columns=['사고여부', '내차피해', '타차가해'])
df.head()
```

## 튜닝여부 더미 변수 변환

```python
# 튜닝여부를 0과 1로 변환 (튜닝X - 0 / 튜닝O - 1)
cond = df['튜닝여부'].eq(0)
df['튜닝여부'] = np.where(cond, 'N', 'Y')
df.head()
```
