## 색상 통일

```python
df['색상'].unique()
```

```python
color_g = ['은색', '은색투톤', '쥐색', '명은색', '은회색', '연금색', '회색']
color_w = ['흰색', '흰색투톤']
color_b = ['검정투톤', '검정색', '은하색']
color_c = ['담녹색', '진주색', '노란색', '청색', '빨간색', '진주투톤', '하늘색', '금색', '분홍색', '주황색', '갈대색', '청옥색', '갈색', '자주색', '연두색', '보라색', '녹색', '갈색투톤', '기타']
```

```python
df.loc[df['색상'].isin(color_g), '색상'] = '은색'
df.loc[df['색상'].isin(color_w), '색상'] = '흰색'
df.loc[df['색상'].isin(color_b), '색상'] = '검정'
df.loc[df['색상'].isin(color_c), '색상'] = '기타'
```
