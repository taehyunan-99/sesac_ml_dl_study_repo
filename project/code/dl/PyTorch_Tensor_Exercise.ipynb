```python
# ts1의 원소 자료형을 확인하세요.
ts1.dtype
```

### 연습 문제 3: 배치 차원 추가 및 제거

한 장의 흑백 이미지가 (높이, 너비) = (28, 28) 형태로 들어있다고 가정합니다.

```python
# 시드를 고정합니다.
torch.manual_seed(seed = 1)
```

```python
# 흑백 이미지 텐서를 생성합니다.
img = torch.randn(size = (28, 28))
```

```python
# img를 확인합니다.
img
```

```python
# img에 채널 차원을 추가하여 (채널, 높이, 너비) = (1, 28, 28) 형태인 img_a를 생성하세요.
img_a = img.unsqueeze(dim=0)
```

```python
# img_a의 형태를 확인하세요.
img_a.shape
```

```python
# img_a에 배치 차원을 추가하여 (배치, 채널, 높이, 너비) = (1, 1, 28, 28) 형태인 img_b를 생성하세요.
img_b = img_a.unsqueeze(dim=0)
```

```python
# img_b의 형태를 확인하세요.
img_b.shape
```

```python
# img_b에서 배치 차원을 제거하여 (채널, 높이, 너비) = (1, 28, 28) 형태인 img_c를 생성하세요.
img_c = img_b.squeeze(dim=0)
```

```python
# img_c의 형태를 확인하세요.
img_c.shape
```

### 연습 문제 4: 완전연결층 입력을 위한 평탄화

다음 텐서는 32장의 컬러 이미지(3, 32, 32)로 이루어진 미니배치입니다.

```python
# 시드를 고정합니다.
torch.manual_seed(seed = 1)
```

```python
# 미니배치를 생성합니다.
# [참고] 텐서 x의 형태를 (배치, 채널, 높이, 너비)이라고 가정합니다.
x = torch.randn(size = (32, 3, 32, 32))
```

```python
# x를 확인합니다.
x
```

```python
# x를 완전연결층에 입력할 수 있도록 (32, 3*32*32) 형태인 x_flat을 생성하세요.
x_flat = x.reshape(32, -1)
```

```python
# x_flat의 형태를 확인하세요.
x_flat.shape
```

### 연습 문제 5: 차원 순서 바꾸기(NHWC -> NCHW)

어떤 라이브러리에서 이미지를 (배치, 높이, 너비, 채널) 구조로 반환한다고 가정합니다.

```python
# 시드를 고정합니다.
torch.manual_seed(seed = 1)
```

```python
# 미니배치를 생성합니다.
# [참고] 텐서 x의 형태를 (배치, 높이, 너비, 채널)이라고 가정합니다.
x = torch.randn(size = (32, 32, 32, 3))
```

```python
# x를 확인합니다.
x
```

```python
# x를 PyTorch CNN 입력 형태인 (N, C, H, W)로 변환한 결과를 확인하세요.(permute 메서드 사용)
x.permute(0, 3, 1, 2)
```

```python
# x를 PyTorch CNN 입력 형태인 (N, C, H, W)로 변환한 결과를 확인하세요.(transpose 메서드 사용)
x.transpose(1, 2).transpose(1, 3)
```
