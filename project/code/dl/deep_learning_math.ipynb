# 딥러닝을 위한 기초 수학

## 함수의 합성

```python
f1 = lambda x: 2 * x + 1
f2 = lambda x: 3 * x - 2

x = 3
f2(f1(x)) # 19
```

```python
f1 = lambda x1, x2, x3: 2 * x1 + 1.5 * x2 + (-1.2) * x3 + 0.5
f2 = lambda x1: 3 * x1 - 1.2

x1, x2, x3 = 1, 1, 0
f2(f1(x1, x2, x3)) # 10.8
```

## 미분 : 변화률과 민감도

```python
def f(x):
    return x ** 2
```

```python
x, h = 3, 1e-5

(f(x + h) - f(x)) / h
# 6.000009999951316
```

## 편미분의 직관적 이해

```python
x1, x2 = 2.0, 3.0
w1, w2 = 0.5, -1.0

w1 * x1 + w2 * x2 # -2.0
(w1 + 0.1) * x1 + w2 * x2 # -1.8
```

## 점곱

### 배열의 점곱

```python
import numpy as np

w = np.array([0.5, -1.0])
x = np.array([2.0, 3.0])

np.dot(w, x).item() # -2.0
```

### 텐서의 점곱

```python
import torch

w = torch.tensor([0.5, -1.0])
x = torch.tensor([2.0, 3.0])

torch.dot(w, x).item() # -2.0
```

## 활성화 함수의 수학적 역할

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.array([-2.0, 1.0, 3.0])
relu(x) # array([0., 1., 3.])
```

## 소프트맥스

```python
z = np.array([2.0, 1.0, -0.1])

z = z - np.max(z)
exp_z = np.exp(z)
softmax = exp_z / np.sum(exp_z)

softmax # array([0.67098969, 0.24684331, 0.082167  ])
np.sum(softmax) # 1.0
```
