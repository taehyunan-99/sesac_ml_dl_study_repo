## 1차원 텐서 생성

```python
li1 = [1, 2, 3]
# li1 = [1, 2.0, 3]
# li1 = [1, 2.0, '3']

ar1 = np.array(li1)
# array([1, 2, 3])
# array([1., 2., 3.])
# array(['1', '2.0', '3'], dtype='<U32')

ts1 = torch.tensor(li1)
# tensor([1, 2, 3])
# tensor([1., 2., 3.])
# 오류 - 파이토치 텐서는 수치형 원소만 가질 수 있음
```

```python
type(ts1)
# torch.Tensor
ts1.ndim
# 1
ts1.shape
# torch.Size([3])
ts1.numel()
# 3
ts1.dtype
# torch.int64
```

## 1차원 텐서 원소의 자료형 변환

```python
ts1.to(torch.float)
# tensor([1., 2., 3.])
ts1.to(torch.float).dtype
# torch.float32
```

```python
ts1 = ts1.float()
ts1.dtype
# torch.float32
```

```python
ts1.to(torch.long)
# tensor([1, 2, 3])
ts1.to(torch.long).dtype
# torch.int64
```

```python
ts1 = ts1.long()
ts1.dtype
# torch.int64
```

## 차원 추가 및 제거

- squeeze 메서드는 원소 개수가 1인 축만 제거할 수 있음
    - dim 지정 생략 시 원소 개수가 1인 축을 모두 제거

```python
ts1.unsqueeze(dim=0)
# tensor([[0, 1, 2]])
ts1.unsqueeze(dim=0).shape
# torch.Size([1, 3])
```

```python
ts2.squeeze(dim=1)
# tensor([0, 1, 2])
ts2.squeeze(dim=1).shape
# torch.Size([3])
```

## 텐서의 결합

```python
x1 = torch.tensor([0, 1, 2])
x2 = torch.tensor([3, 4, 5])

torch.stack(tensors=[x1, x2], dim=0)
# tensor([[0, 1, 2],
#         [3, 4, 5]])
torch.stack(tensors=[x1, x2], dim=1)
# tensor([[0, 3],
#         [1, 4],
#         [2, 5]])
```

```python
gray_img_1 = torch.randn(1, 28, 28)
gray_img_2 = torch.randn(1, 28, 28)

img_batch_1 = torch.stack(tensors=[gray_img_1, gray_img_2], dim=0)
img_batch_1.shape
# torch.Size([2, 1, 28, 28])
```

## 텐서 입력 형태 변환

```python
ts1
# tensor([0, 1, 2])
```

```python
ts1 = ts1.unsqueeze(dim=0).unsqueeze(dim=0)
ts1.shape
# torch.Size([1, 1, 3])
```

- permute 메서드는 차원 축을 일괄로 바꾸는 기능이 있음
    - np.transpose() 가 동일한 기능을 수행

```python
ts1.permute(0, 2, 1).shape
# torch.Size([1, 3, 1])
ts1.permute(2, 0, 1).shape
# torch.Size([3, 1, 1])
```

- transpose 메서드는 2개의 차원을 서로 맞바꾸는 기능이 있음
    - np.swapaxes() 가 동일한 기능을 수행

```python
ts1.transpose(1, 2).shape
# torch.Size([1, 3, 1])
ts1.transpose(0, 2).shape
# torch.Size([3, 1, 1])
```
